```python
# 2.Sending the OTP to user's E-mail address
import smtplib
import random
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# Creating a function to generate OTP
def otp_generation():
    return ''.join(random.choices([str(i) for i in range(10)], k=6))

# Calling the function
otp = otp_generation()

# Email details
sender_email = input("please Enter the Sende Email:")
receiver_email = input("please Enter the receiver Email:")
password = input("Enter the password:")

# Creating the email message
subject = "Your OTP Code"
body = f"Your OTP code is {otp}."

msg = MIMEMultipart()
msg['From'] = sender_email
msg['To'] = receiver_email
msg['Subject'] = subject

msg.attach(MIMEText(body, 'plain'))

try:
    # Setting up the SMTP server
    server = smtplib.SMTP("smtp.gmail.com", 587)
    server.starttls()
    server.login(sender_email, password)

    # Sending the email
    server.sendmail(sender_email, receiver_email, msg.as_string())
    print("OTP has been sent to", receiver_email)
except Exception as e:
    print(f"Failed to send email: {e}")
finally:
    # Closing the server connection
    server.quit()
```

```python
# libraries required
import smtplib
import random
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from tkinter import *
from tkinter import messagebox

# Function to generate OTP
def otp_generation():
    return ''.join(random.choices([str(i) for i in range(10)], k=6))

# Function to send OTP email
def send_otp():
    global otp
    otp = otp_generation()
    
# Email details
    sender_email = "hemalathachilaka@gmail.com"
    receiver_email = email_entry.get()
    password =input("Enter the Password: ")


# Creating the email message
    subject = "Your OTP Code"
    body = f"Your OTP code is {otp}."

    msg = MIMEMultipart()
    msg['From'] = sender_email
    msg['To'] = receiver_email
    msg['Subject'] = subject

    msg.attach(MIMEText(body, 'plain'))

    try:
# Setting up the SMTP server
        server = smtplib.SMTP("smtp.gmail.com", 587)
        server.starttls()
        server.login(sender_email, password)

# Sending the email
        server.sendmail(sender_email, receiver_email, msg.as_string())
        result_label.config(text="OTP has been sent")
        
# Hide email frame and show OTP frame
        email_frame.pack_forget()
        otp_frame.pack(pady=10)
    except Exception as e:
        print(f"Failed to send email: {e}")
        messagebox.showerror("Error", "Failed to send email. Please try again.")
    finally:
# Closing the server connection
        server.quit()

# Function to check OTP entered in the GUI
def check_otp():
    global attempts
    entered_otp = otp_entry.get()
    if entered_otp == otp:
        result_label.config(text="ACCESS GRANTED")
        otp_entry.config(state="disabled")
    else:
        attempts -= 1
        if attempts > 0:
            result_label.config(text=f"INCORRECT OTP. You have {attempts} attempts left.")
        else:
            result_label.config(text="NO ATTEMPTS LEFT. ACCESS DENIED")
            messagebox.showerror("Error", "NO ATTEMPTS LEFT. ACCESS DENIED")
            otp_entry.config(state="disabled")
            check_otp_button.config(state="disabled")

# Validation function for OTP entry
def validate_otp(P):
    return P.isdigit() and len(P) <= 6

# Tkinter GUI for Email and OTP input
window = Tk()
window.title("Email and OTP Verification")
window.minsize(width=400, height=200)

# Initialize attempt counter
attempts = 3

# Email frame
email_frame = Frame(window)
email_frame.pack(pady=10)

# Create and place the email label
email_label = Label(email_frame, text="Enter your Email ID:", bd=10)
email_label.grid(row=0, column=0, padx=10, pady=10)

# Create and place the email entry widget
email_entry = Entry(email_frame, width=30, bd=5)
email_entry.grid(row=0, column=1, padx=10, pady=10)

# Create and place the send OTP button
send_otp_button = Button(email_frame, text="Send OTP", command=send_otp)
send_otp_button.grid(row=1, column=1, padx=10, pady=10)

# OTP frame (initially hidden)
otp_frame = Frame(window)
otp_frame.pack_forget()

# Create and place the OTP label
otp_label = Label(otp_frame, text="Please Enter the OTP", bd=10)
otp_label.grid(row=0, column=0, padx=10, pady=10)

# Validate only digits and up to 6 characters for OTP entry
vcmd = (window.register(validate_otp), '%P')

# Create and place the OTP entry widget
otp_entry = Entry(otp_frame, width=20, bd=5, validate='key', validatecommand=vcmd)
otp_entry.grid(row=0, column=1, padx=10, pady=10)

# Create and place the check OTP button
check_otp_button = Button(otp_frame, text="Enter", command=check_otp)
check_otp_button.grid(row=1, column=1, padx=10, pady=10)

# Create and place the result label
result_label = Label(otp_frame, text="", bd=10)
result_label.grid(row=2, column=0, columnspan=2, padx=10, pady=10)

# Run the main event loop to keep the window open
window.mainloop()
```
